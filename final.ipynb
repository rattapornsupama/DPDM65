#Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

X คือ feature ที่เราสนใจ

y คือ sell_price

## กำหนด Xtrain,Ytrain 

```python
Xtrain = data1[['usable_area','number_of_rai','number_of_ngan','number_of_va','number_of_bedroom','number_of_bathroom','number_of_parking']]
ytrain = data1['sell_price']
Xtrain
```

##ใช้คำสั่ง Linear Regression

LR = LinearRegression().fit(Xtrain, ytrain) 

เราเรียกฟังก์ชัน LinearRegression() แล้วใช้ Method .fit ป้อน Vector ของ feature และ Vector ของ label ลงไปในฟังก์ชัน 

โดยฟังก์ชันจะใช้ข้อมูลเหล่านี้มาคำนวนหา Parameter ที่จะลด Cost function ให้มีค่าต่ำที่สุด โดย Parameter สำหรับ Linear regression ก็คือ Coefficient (= w) และ Intercept (= b) อยู่ใน Hypothesis function ของ Multiple Linear regression:

y =w1(x1)+w2(x2)+w3(x3)+w4(x4)+w5(x5)+w6(x6)+w7(x7)+b     



```python
LR = LinearRegression().fit(Xtrain, ytrain)
print("LR coefficient is", LR.coef_)
print("LR intercept is", LR.intercept_)
```

##ใช้คำสั่ง .predict
เราใช้คำสั่ง .predict โดยใส่ข้อมูล datatest1 ลงไป เพื่อให้โมเดลพยากรณ์ค่า y (sell_price) ออกมา

```python
predictions = LR.predict(datatest1)
predictions
```

##ตารางแสดงผลการทำนายราคา 

โดยที่คอลัมน์ชื่อ predictions จะแสดง sell_price ที่เราทำนายโดยโมเดลนี้ จากข้อมูลที่กำหนดให้ 

```python
result = pd.concat([datatest, pd.Series(predictions,name='predictions')],axis='columns')
result
```
