#เตรียมข้อมูล

เปลี่ยน error ทั้งหมดใน แต่ละคอลัมน์ที่เราที่เราสนใจ ให้เป็น ตัวเลข






```python
data['usable_area'] = pd.to_numeric(data['usable_area'],errors='coerce')
data['number_of_story'] = pd.to_numeric(data['number_of_story'],errors='coerce')
data['number_of_bedroom'] = pd.to_numeric(data['number_of_bedroom'],errors='coerce')
data['number_of_bathroom'] = pd.to_numeric(data['number_of_bedroom'],errors='coerce')
data['number_of_parking'] = pd.to_numeric(data['number_of_bedroom'],errors='coerce')
#data['sell_price'] = pd.to_numeric(data['sell_price'],errors='coerce')
```

สร้างตารางข้อมูลใหม่ โดยเอามาเฉพาะคอลัมน์ที่เราสนใจ

```python
data1=data[['usable_area','number_of_story','number_of_bedroom','number_of_bathroom','number_of_parking','sell_price']]
data1
```

เติมค่า missing ด้วย 0

```python
data1 = data1.fillna(0)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

X คือ feature ที่เราสนใจ

y คือ sell_price

```python
X = data1[['usable_area','number_of_story','number_of_bedroom','number_of_bathroom','number_of_parking']]
y = data1['sell_price']
```

```python
lrm = LinearRegression()
lrm.fit(X, y)
```

```python
predictions = lrm.predict(X)
predictions
```

```python
result_data = pd.concat([data1.reset_index(), pd.Series(predictions,name='predictions')],axis='columns')
result_data
```

```python

from sklearn.metrics import mean_squared_error
  
# Calculation of Mean Squared Error (MSE)
mean_squared_error(y,predictions)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y, predictions)
```
